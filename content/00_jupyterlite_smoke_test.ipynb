# JupyterLite Smoke Test

This notebook is the first browser-only test for the course.

It intentionally uses only `numpy` and `matplotlib`, so it is a safe first check that:

- the JupyterLite site builds on GitHub Pages
- the `xeus-python` kernel starts in the browser
- basic numerical work and plotting are available


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0, 2.0 * np.pi, 300)
dx = x[1] - x[0]

f = np.sin(x)
df_exact = np.cos(x)
df_numeric = np.gradient(f, dx)

rms_error = np.sqrt(np.mean((df_numeric - df_exact) ** 2))
print(f"RMS error for d/dx sin(x) ≈ cos(x): {rms_error:.6f}")
```

```python
plt.figure(figsize=(9, 5))
plt.plot(x, df_exact, label="exact derivative")
plt.plot(x, df_numeric, "--", label="numerical derivative")
plt.xlabel("x")
plt.ylabel("amplitude")
plt.title("Minimal finite-difference style test in JupyterLite")
plt.grid(True)
plt.legend()
plt.show()
```
